```python
# Load the dataset from drive

from google.colab import drive

drive.mount("/content/gdrive")

import pandas as pd

df = pd.read_excel("/content/gdrive/MyDrive/Data/UN_USER/20231009_Result_BOS_UNINFO_user_list.xlsx", sheet_name = "BOS_UNINFO_user_list_key_roles")
df['UNINFO.User Roles'].fillna("", inplace=True)
df.head(3)
```

```python
df[df["BOS_AF_user_info.COUNTRY"].str.contains("Tanzania")]
```

```python
df["BOS_AF_user_info.COUNTRY"] = df["BOS_AF_user_info.COUNTRY"].str.replace("Cote DIvoire", "Cote d'Ivoire")
df["BOS_AF_user_info.COUNTRY"] = df["BOS_AF_user_info.COUNTRY"].str.replace("United Republic of Tanzania", "Tanzania")
```

```python
df["BOS_AF_user_info.COUNTRY"].unique()
```

```python
# Filter the position based on accoutnability

title_to_check = "OMT Chair"

df_filtered = df[df["BOS_AF_user_info.BOS_ACCOUNTABILITY"] == title_to_check]
df_filtered.sort_values(by="BOS_AF_user_info.COUNTRY")

print(df_filtered["BOS_AF_user_info.BOS_ACCOUNTABILITY"].value_counts(), "\n") #Yes! Only OMT chair is in the file
df_filtered = df_filtered.sort_values(by="BOS_AF_user_info.COUNTRY")
```

```python
print(len(df_filtered["BOS_AF_user_info.COUNTRY"].unique()))

df_filtered["BOS_AF_user_info.COUNTRY"].unique()
```

```python
# Extract the columns which is needed

df_result = df_filtered[["BOS_AF_user_info.COUNTRY", "First_Name", "Last_Name", "Email", "BOS_AF_user_info.AGENCY", "UNINFO.email"]].sort_values(by="BOS_AF_user_info.COUNTRY")
df_result
```

```python
df_result.fillna("None", inplace=True)
df_result["Existence_UNINFO"] = df_result["UNINFO.email"] != "None"

df_result
```

```python
df_count = df_result.groupby("BOS_AF_user_info.COUNTRY").agg({
    "First_Name": "count",
    "Existence_UNINFO":"sum",
}).reset_index()

df_count.rename(columns = {"First_Name":"N_BOS_Users", "Existence_UNINFO":"N_UNINFO_Users"}, inplace=True)

df_count
```

```python
# merge the dataset

df_merged = pd.merge(df_count, df_result, on =["BOS_AF_user_info.COUNTRY"] )

df_merged
```

```python
# Rename Columns

df_merged.rename(columns={"BOS_AF_user_info.COUNTRY": "Country", "BOS_AF_user_info.AGENCY": "Agency"}, inplace=True)

df_merged
```

```python
df_merged[df_merged["Country"].str.contains("Tan")]
```

```python
# Add region information

df_region = pd.read_excel("/content/gdrive/MyDrive/Data/UN_USER/20231010_BOS_UNINFO_priority_users.xlsx",
                           sheet_name = "1.b. Heads of RCO-Details")

df_region = df_region[["Region", "Country"]]

print(df_region[df_region["Country"].str.contains("Tanza")])
print(df_region[df_region["Country"].str.contains("Cote")])
```

```python
df_region_merged  = pd.merge(df_region, df_merged, on =["Country"], how = "outer")

df_region_merged["Region"].fillna("None", inplace=True)

df_region_merged[df_region_merged["Region"] == "None"]

len(df_region_merged["Country"].unique())
```
